For this problem we’ll be working with two years of the bicycle sharing systems for New York City (Citibike). The dataset contains daily bike trip counts, along with daily measurements on environmental and seasonal information that may affect the bikesharing.

Here’s information on what the variables mean.

- trips - daily total number of bike trips taken (all stations)
- precipitation - daily inches of rain
- snow_depth - daily inches of snow. Accoding to NOAA "Determine the depth of the new and old snow remaining on the ground at observation time".
- snowfall - according to NOAA "Measure and record the snowfall (snow, ice pellets ) since the previous snowfall observation (24 hours)."
- max_temperature - daily maximum temperature in Farenheit (highest temperature reached)
- min_temperature - daily minimum temperature in Farenheit (lowest temperature reached)
- average_wind_speed - measured hourly in mph and averaged for daily value
- year 
- holiday
    - True
    - False
- stations in service - docking stations working per day
- weekday
    - True
    - False
- weekday_non_holiday
    - True
    - False

More details on snow measurements [here](https://www.weather.gov/gsp/snow)

### 1. Data pre-processing [8 pts]
Let's start by loading and pre-processing our dataset.

**1)a) Load the `bikes` dataset into a dataframe called `bikes`, and print the first five rows.** [1pt]

```python
# your code here
bikes = pd.read_csv('daily_citi_bike_trip_counts_and_weather.csv')
print(bikes.head(5))
```

**1)b) Check that there are no missing values in your dataset. If they are, take care of them appropriately.** [1pt]


```python
# Your code here
bikes = bikes.dropna()
```

**1)c) Separate your dataset into two dataframe. One dataframe should contain the data for weekends and the other one for weekdays.** [2pt]

```python
# Your code here
weekends = bikes.loc[bikes['weekday'] == False]
weekdays = bikes.loc[bikes['weekday'] == True]
```

**1)d) Let's look at the number of bicycle rental (trips) per season! Draw a boxplot that displays this information for the whole dataset. As subplots, draw the boxplots for weekdays and weekends. Is there a differece in the distribution of trips between seasons? Is there a difference between weekdays and weekends? Explain.** [4pts]

```python
# Your code here
fig, axes = plt.subplots(3, 1, figsize=(5, 15))
sns.boxplot(ax = axes[0],
            data = bikes,
            x = 'season',
            y = 'trips',)
axes[0].set_title('Total trip by season')

sns.boxplot(ax = axes[1],
            data = weekends,
            x = 'season',
            y = 'trips')
axes[1].set_title("Weekends trip by season")

sns.boxplot(ax = axes[2],
            data = weekdays,
            x = 'season',
            y = 'trips')
axes[2].set_title('Weekdays trip by season')

plt.show()
```

## Problem 3: Multiple linear regression [8 pts]

In this problem we’ll practice fitting and interpreting the results of a multiple linear regression. For this question use the weekdays dataset.

**3)a) Fit a regression model with trips as the target variable and the following variables as features: snow_depth, snowfall, min_temperature, max_temperature, precipitation, month** [2pt]


```python
# Your code here
mod3 = smf.ols('trips ~ snow_depth+snowfall+min_temperature+max_temperature+precipitation+month', data=weekdays)
lm3 = mod3.fit()
print(lm3.summary())
```

**3)b) Interpret the coefficients of snow_depth, snowfall, min_temperature, max_temperature, precipitation, month in the model.** [2pt]

--> *Your answer here*
Each unit increase for snow_death will decrease the trip by 280, each unit increase for snowfall will decrease the bike trips by 71, each unit increase in minimum temperature will decrease the bike trip by 11, each unit increase in max_temperature will increase the bike trip by 460, each unit increase in precipitation will decrease the bike trip by 8226, each unit increase in month will increase trip by 661.

**3)c) Re-run the model from 3)a) but add another predictor of your choice. Justify your choice of predictor. Did this make any difference? Did the R-squared of the model improve?** [2pt]

```python
# Your code here
mod4 = smf.ols('trips ~ snow_depth+snowfall+min_temperature+max_temperature+precipitation+month+average_wind_speed', data=weekdays)
lm4 = mod4.fit()
print(lm4.summary())
```

--> *Your answer here*
I added one more predictor of average wind speed into the fitted model, since I think the higher windspeed will also contribute to a loss of bike trips, since people don't like heavy wind during riding bikes. The results shows similar outcome, for each unit increase of average wind speed, the bike trip will decrease by 349. The R^2 of the fitted model also increase and improve a little bit.

**5)b) Try out six different degrees of polynomial regression fits for modeling the relationship between trips and month. Do not do more than six models. Display all of them in the same plot (through subplots). Choose the subplot that appears to nicely capture the trends in the data. Be sure to print the $R^2$ or adjusted-$R^2$ for each subplot to help you with your decision. Explain your choice and include comments to explain your code.** [8pts]

```python
# Your code here
fig, axes = plt.subplots(3, 2)
x=np.array(weekdays.loc[:,'month']).reshape(-1,1)
y=np.array(weekdays.loc[:,'trips']).reshape(-1,1)

poly2 = PolynomialFeatures(degree = 2)
X_poly2 = poly2.fit_transform(x)
poly2.fit(X_poly2, y)
lin2 = LinearRegression()
lin2.fit(X_poly2, y)
y_pred2 = lin2.predict(X_poly2)
r2_2 = r2_score(y, y_pred2)
print('R squared for degree 2 is', r2_2)

poly3 = PolynomialFeatures(degree = 3)
X_poly3 = poly3.fit_transform(x)
poly3.fit(X_poly3, y)
lin3 = LinearRegression()
lin3.fit(X_poly3, y)
y_pred3 = lin3.predict(X_poly3)
r2_3 = r2_score(y, y_pred3)
print('R squared for degree 3 is', r2_3)

poly4 = PolynomialFeatures(degree = 4)
X_poly4 = poly4.fit_transform(x)
poly4.fit(X_poly4, y)
lin4 = LinearRegression()
lin4.fit(X_poly4, y)
y_pred4 = lin4.predict(X_poly4)
r2_4 = r2_score(y, y_pred4)
print('R squared for degree 4 is', r2_4)

poly5 = PolynomialFeatures(degree = 5)
X_poly5 = poly5.fit_transform(x)
poly5.fit(X_poly5, y)
lin5 = LinearRegression()
lin5.fit(X_poly5, y)
y_pred5 = lin5.predict(X_poly5)
r2_5 = r2_score(y, y_pred5)
print('R squared for degree 5 is', r2_5)

poly6 = PolynomialFeatures(degree = 6)
X_poly6 = poly6.fit_transform(x)
poly6.fit(X_poly6, y)
lin6 = LinearRegression()
lin6.fit(X_poly6, y)
y_pred6 = lin6.predict(X_poly6)
r2_6 = r2_score(y, y_pred6)
print('R squared for degree 6 is', r2_6)

poly7 = PolynomialFeatures(degree = 7)
X_poly7 = poly7.fit_transform(x)
poly7.fit(X_poly7, y)
lin7 = LinearRegression()
lin7.fit(X_poly7, y)
y_pred7 = lin7.predict(X_poly7)
r2_7 = r2_score(y, y_pred7)
print('R squared for degree 7 is', r2_7)


sns.regplot(x=x, y=y, order=2, ax=axes[0,0], scatter_kws={"color": "grey"}, line_kws={"color": "red"})
sns.regplot(x=x, y=y, order=3, ax=axes[0,1], scatter_kws={"color": "grey"}, line_kws={"color": "red"})
sns.regplot(x=x, y=y, order=4, ax=axes[1,0], scatter_kws={"color": "grey"}, line_kws={"color": "red"})
sns.regplot(x=x, y=y, order=5, ax=axes[1,1], scatter_kws={"color": "grey"}, line_kws={"color": "red"})
sns.regplot(x=x, y=y, order=6, ax=axes[2,0], scatter_kws={"color": "grey"}, line_kws={"color": "red"})
sns.regplot(x=x, y=y, order=7, ax=axes[2,1], scatter_kws={"color": "grey"}, line_kws={"color": "red"})

plt.show()
```

|--> *Your answer here*
Based on the graphs and the R squared, I am able to say that degree 5 seems like the most fitted, (not overfitted or underfitted), polynomial regression for this dataset, since the curve seems not overfitting, and also the R sqaured is still maintaining a relatively high score comparing to others.
